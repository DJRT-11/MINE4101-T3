```python
import pandas as pd
import numpy as np
import random
```

```python
churn_df = pd.read_json("Model Training\data\churn_future.json")
churn_df
```

```python
churn_bsl = churn_df.sample(frac=0.5, random_state=1)
churn_opt = churn_df.loc[~churn_df.index.isin(churn_bsl.index)]
```

```python
print(churn_bsl.shape)
print(churn_opt.shape)
```

```python
churn_bsl['week_day'] = [random.randrange(7) for _ in range(len(churn_bsl))]
churn_opt['week_day'] = [random.randrange(7) for _ in range(len(churn_opt))]

day_map = {0: 'MON', 
           1: 'TUE', 
           2: 'WED', 
           3: 'THU', 
           4: 'FRI', 
           5: 'SAT', 
           6: 'SUN'}

churn_bsl['week_day'] = churn_bsl['week_day'].map(day_map)
churn_opt['week_day'] = churn_opt['week_day'].map(day_map)
```

```python
print(churn_bsl.value_counts("week_day"))
print(churn_opt.value_counts("week_day"))
```

```python
export_df = churn_bsl.loc[churn_bsl["week_day"] == "SUN"].drop(['week_day'], axis=1)
export_df.to_json("./Model Training/data/churn_future_divs/churn_future_baseline_SUN.json", orient="records")
```

```python
export_df = churn_opt.loc[churn_opt["week_day"] == "SUN"].drop(['week_day'], axis=1)
export_df.to_json("./Model Training/data/churn_future_divs/churn_future_model_SUN.json", orient="records")
```
